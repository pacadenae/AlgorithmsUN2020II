```
import matplotlib.pyplot as plt
import numpy as np
from time import time
```

```
def recur_fibo(n):
  if n = 0:
    return 0
  if n <= 1:
      return n
  else:
      return (recur_fibo(n-1) + recur_fibo(n-2))
```

```
def ite_fibo(n):
    a= 0
    b = 1
    for i in range(n):
        a= b
        b = a + b
    return a
```

```
allTime = []
y= []

for i in range(50):
  start_time = time()
  print(i, recur_fibo(i))
  y.append(i)
  timeT = time()-start_time
  allTime.append(timeT)
  print(timeT) 

plt.plot(y, allTime)


```
